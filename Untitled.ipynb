```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm,zscore,pearsonr
import math

df=pd.read_csv("Indian_cities.csv")
def check_outlier(array):
    q1=np.percentile(array,25)
    q3=np.percentile(array,75)
    lower=q1-((q3-q1)*1.5)
    upper=q3+((q3-q1)*1.5)
    return list(filter(lambda x:lower<=x<=upper,array))
def check_normal(data):
    p=list(map(lambda x:((x-0.5)/len(data)),data))
    z=zscore(p)
    plt.scatter(data,z)
    plt.plot()
    plt.show()
def normalize(data):
    mu=np.mean(data)
    sd=np.std(data)
    a=[]  
    raw_wo_outlier=check_outlier(data)
    for i in data:
       a.append((i-mu)/sd) 
    print(mu,sd)
    #for i in a:
     #   print(i)
    
    a.sort()
    b=check_outlier(a)
    print(len(b))
    #print(a)
    plt.plot(b,norm.pdf(b,0,1))
    #plt.xticks(np.arange(-1.5,1.5))
    plt.show()
    check_normal(raw_wo_outlier)
    
    
    
normalize(df.population_total)
normalize(df.population_male)
normalize(df.population_female)
normalize(df.population_total_0_to_6)
normalize(df.sex_ratio)
normalize(df.effective_literacy_rate_total)
sns.set(color_codes=True)
sns.distplot(df['effective_literacy_rate_total'])

normalize(df.effective_literacy_rate_male)
normalize(df.effective_literacy_rate_female)

```

```python
def histogram(x):
    p=[]
    for i in x:
        if i not in p:
            p.append(i)
    counter=[]
    for i in p:
        count=0
        for j in x:
            if i==j:
                count+=1
        counter.append(count)
    print(p)
    print(counter)
    sns.barplot(x=p,y=counter,data=df)
    plt.xticks(rotation=90)
    plt.xlabel('States')
    plt.ylabel('no. of muncipal cities')
    plt.show()
histogram(df.state_name)
import seaborn as sns

```

```python
def stacked_bar(n,male,female,city):
    a=np.arange(n)
    plt.bar(a,male,label='male',color='blue')
    plt.bar(a,female,bottom=male,label='female',color='green')
    plt.legend(loc='upper right')
    plt.xticks(a,city,rotation=90)
    plt.title('Number of graduates')
    plt.ylabel('Graduates')
    plt.show()
```

```python
def check(x,y):
    a=x.sample(n=30)
    b=y.sample(n=30)
    x1_bar=np.mean(a)
    x2_bar=np.mean(b)
    s1=np.std(a)
    s2=np.std(b)
    n1=len(a)
    n2=len(b)
    print(x1_bar,x2_bar,n1,n2,s1,s2)
    z=(x1_bar-x2_bar)/math.sqrt((pow(s1,2)/n1)+(pow(s2,2)/n2))
    print(z)
    p=1-norm.cdf(z)
    print(p)
    if(p<0.05):
        print("null hypothesis rejected")
    else:
        print("null hypothesis plausible")
check(df.effective_literacy_rate_male,df.effective_literacy_rate_female)
check(df.male_graduates,df.female_graduates)
check(df.population_male,df.population_female)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm,zscore,pearsonr
import math

df=pd.read_csv("Indian_cities.csv")
#import seaborn as sn
#plt.figure(figsize=(30,7))
#%matplotlib inline
def check_outlier(array):
    q1=np.percentile(array,25)
    q3=np.percentile(array,75)
    lower=q1-((q3-q1)*1.5)
    upper=q3+((q3-q1)*1.5)
    return list(filter(lambda x:lower<=x<=upper,array))
def box(n,m):
    x=n.sample(n=100)
    y=m.sample(n=100)
    a=check_outlier(x)
    b=check_outlier(y)

    #sn.boxplot(x=b,data=df)
    data_to_plot = [a,b]
    fig = plt.figure(1, figsize=(9, 6))


    # Create an axes instance
    ax = fig.add_subplot(111)

    # Create the boxplot
    bp = ax.boxplot(data_to_plot,patch_artist=True)
    for box in bp['boxes']:
        # change outline color
        box.set( color='green', linewidth=2)
        # change fill color
        box.set( facecolor = 'red' )
    for whisker in bp['whiskers']:
        whisker.set(color='#7570b3', linewidth=2)
    for cap in bp['caps']:
        cap.set(color='#7570b3', linewidth=2)
    for median in bp['medians']:
        median.set(color='#b2df8a', linewidth=2)
    for flier in bp['fliers']:
        flier.set(marker='o', color='#e7298a', alpha=0.5)
    ax.set_xticklabels(['Male literates','Female literates'])
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    plt.title('total literates')
    plt.ylabel('population')
    # Save the figure
    #fig.savefig('fig2.png', bbox_inches='tight')
box(df.literates_male,df.literates_female)
```
